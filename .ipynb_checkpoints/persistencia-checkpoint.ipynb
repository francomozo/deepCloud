```python
import os
import matplotlib.pyplot as plt
import cv2
import re
import numpy as np
```

```python
def load_images_from_folder(folder,numeric_name = False):
    '''
    Input:
    folder: path y nombre de la carpeta con las imagenes 
    numeric_name: borrar letras de los nombres de las imagenes al cargar a la lista
    Output:
    images: lista con las imagenes cargadas
    img_names: lista con el nombre de los archivos sin valores letras, solo numeros'''
    images = []
    img_name = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img[:,:,0])
            if (numeric_name):
                img_name.append(re.sub("[^0-9]", "",filename))
            else:
                img_name.append(filename)
    return images,img_name
```

```python
images_20,img_name = load_images_from_folder('imagenes/dia_20', True)
images_91,img_name = load_images_from_folder('imagenes/dia_91', True)
images_277,img_name = load_images_from_folder('imagenes/dia_277', True)
images_305,img_name = load_images_from_folder('imagenes/dia_305', True)
```

```python
prediccion20 = images_20[70]
prediccion91 = images_91[70]
prediccion277 = images_277[70]
prediccion305 = images_305[70]
errores20 = []
errores91 = []
errores277 = []
errores305 = []
for i in range(71, 100):
    diferencia20 = np.sqrt(np.mean((images_20[i] - prediccion20)**2))
    diferencia91 = np.sqrt(np.mean((images_91[i] - prediccion91)**2))
    diferencia277 = np.sqrt(np.mean((images_277[i] - prediccion277)**2))
    diferencia305 = np.sqrt(np.mean((images_305[i] - prediccion305)**2))
    errores20.append(diferencia20)
    errores91.append(diferencia91)
    errores277.append(diferencia277)
    errores305.append(diferencia305)
    

tiempo = np.linspace(10,300,29)


plt.plot(tiempo,errores20, 'r.',tiempo,errores91,'b.',tiempo,errores277,'y.',tiempo,errores305,'g.')
plt.xlabel('Tiempo (minutos)') 
plt.ylabel('RMSE') 
plt.title('RMSE persistencia') 
plt.show()
```
